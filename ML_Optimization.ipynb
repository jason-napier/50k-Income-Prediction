```python
def MLTrials(independant_dataframe, dependent_series, name0="0", name1="1"):
    x = independant_dataframe
    y= dependent_series

    import pandas as pd
    import numpy as np
    from sklearn.preprocessing import StandardScaler
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

    #Data Processing
    y = y.ravel()
    y = y.reshape(-1,1)

    x = pd.DataFrame(x)
    x = pd.get_dummies(x)

    # Define function for executing machine models
    def MLModel(x, y, model, model_name=""):

        # Split training data and test data
        from sklearn.model_selection import train_test_split
        X_train, X_test, y_train, y_test = train_test_split(x,
                                                            y,
                                                            random_state=1,
                                                            )
        # Scale Data Process

        # Create StandardScaler Instance
        scaler = StandardScaler()

        # Fit Standard Scaler
        X_scaler = scaler.fit(X_train)

        # Scale Data
        X_train_scaled = X_scaler.transform(X_train)
        X_test_scaled = X_scaler.transform(X_test)

        # Fit and Predict Model
        model.fit(X_train_scaled, y_train)
        predictions = model.predict(X_test_scaled)

        # Calculate the Confusion Matrix
        cm = confusion_matrix(y_test, predictions)

        cm_df = pd.DataFrame(
            cm, index=[f"Actual {name0}", f"Actual {name1}"], columns=[f"Predicted {name0}", f"Predicted {name1}"]
        )

        # Calculate Accuracy Score
        acc_score = accuracy_score(y_test, predictions)

        # Display Results
        print(f"Model Used: {model_name}")
        print(cm_df)
        print(f"Accuracy Score: {acc_score}")

        #Create Classification Report
        print("Classification Report")
        print(classification_report(y_test, predictions,
                                    target_names=[name0,name1]))

    #Run Different Machine Learning Models

    # Logistic Regression Model
    from sklearn.linear_model import LogisticRegression
    model = LogisticRegression(solver='lbfgs', random_state=1)

    MLModel(x,y,model,"Logistic Regression Model")

    # Support Vector Machine
    from sklearn.svm import SVC
    model = SVC(kernel='linear')

    MLModel(x,y,model,"Support Vector Machine")

    # Decision Trees
    from sklearn import tree
    model = tree.DecisionTreeClassifier()

    MLModel(x,y,model,"Decision Trees")

    # Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier()

    MLModel(x,y,model,"Random Forest")

    # K Neighbors
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier(n_neighbors=3)

    MLModel(x,y,model,"K Neighbors")
```
